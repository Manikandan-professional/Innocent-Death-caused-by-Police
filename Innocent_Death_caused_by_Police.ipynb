# **Innocent-Death-caused-by-Police**

**Importing the Necessary Moduels**

```python
import pandas as pd
import numpy as np
```

**Reading the csv file**

```python
a=pd.read_csv('https://raw.githubusercontent.com/Manikandan-professional/Innocent-Death-caused-by-Police/main/Innocent%20Deaths%20caused%20by%20Police%20(All%20time).csv')
a
```

**Conversion of csv file to DataFrame**

```python
df=pd.DataFrame(a)
```

**First five row preview of the loaded dataset**

```python
df.head()
```

**Columns in the DataFrame(df)**

```python
list(df.columns)
```

```python
top_10=pd.DataFrame(df.Name.value_counts().head(10))
top_10
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8),dpi=80)
ax=plt.axes()
ax.set_facecolor('black')
xmin, xmax, ymin, ymax = plt.axis([-500, 1500,0,10])
plt.plot(top_10.Name,top_10.index,color='red')
plt.barh(top_10.index,top_10.Name,color='gold')
plt.scatter(top_10.Name,top_10.index,color='gold')
plt.xticks(rotation=50)
plt.show()
```
